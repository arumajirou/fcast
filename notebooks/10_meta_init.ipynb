# 10 Meta DB init
SQLite（デフォルト）で `forecast_meta` 相当のテーブルを作成し、一覧を確認。

```python
from fcast.meta.db import init_db, get_engine
init_db()
engine = get_engine()
print(engine)
```

```python
from sqlalchemy import inspect
insp = inspect(engine)
tables = sorted(insp.get_table_names())
print("tables:", len(tables))
tables[:20]
```
