# 30 Games & Slots
宝くじゲーム（増減）とスロット（N1..Nk）を定義。

```python
from fcast.meta.db import init_db, session_scope
from fcast.meta.orm import DatasetsGame, DatasetsGameSlot
init_db()
```

```python
games = {
  "numbers3": {"k":3, "maxv":9},
  "numbers4": {"k":4, "maxv":9},
  "miniloto": {"k":5, "maxv":31},
  "loto6": {"k":6, "maxv":43},
  "loto7": {"k":7, "maxv":37},
  "bingo5": {"k":5, "maxv":39},
}
with session_scope() as s:
    for code, cfg in games.items():
        g = s.query(DatasetsGame).filter_by(game_code=code).one_or_none()
        if g is None:
            g = DatasetsGame(game_code=code, rules_json={"k":cfg["k"],"maxv":cfg["maxv"]})
            s.add(g); s.flush()
            for i in range(1, cfg["k"]+1):
                s.add(DatasetsGameSlot(game_id=g.game_id, slot_code=f"N{i}", slot_index=i))
print("done")
```
